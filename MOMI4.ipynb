```python
# importing modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('./datasets/LFR2023_SectionI.xlsx', sheet_name='I4', skiprows=6)
df
```

```python
df = df.iloc[36:54]
df
```

```python
df.rename({"Unnamed: 0": "Industry (SSIC 2020)", "Less Than 1 Year": "<1 Year", "1 Year to Less Than 5\nYears": "1-5 Years", "5 Years to Less Than\n10 Years": "5-10 Years", "10 Years to Less\nThan 20 Years": '10-20 Years', "At Least 20 Years": ">20 Years"} , axis=1 , inplace=True)
```

```python
df.drop(36, inplace=True)
```

```python
numerical_cols = ['Total' ,'<1 Year', '1-5 Years', '5-10 Years', '10-20 Years', '>20 Years']
for col in numerical_cols:
    df[col] = df[col].astype(float) # np.astype to float so that numbers can be rounded

    df[col] = np.array(df[col]) * 1000
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64') # Convert to int while keeping NaN values

df
```
